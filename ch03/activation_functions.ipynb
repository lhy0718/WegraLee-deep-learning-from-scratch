```python
# step function
import matplotlib.pylab as plt
import numpy as np


def step_function(x):
    return np.array(x > 0, dtype=np.int32)


X = np.arange(-5.0, 5.0, 0.1)
Y = step_function(X)
plt.plot(X, Y)
plt.ylim(-0.1, 1.1)  # y축의 범위 지정
plt.show()
```

```python
# sigmoid function
import matplotlib.pylab as plt
import numpy as np


def sigmoid(x):
    return 1 / (1 + np.exp(-x))


X = np.arange(-5.0, 5.0, 0.1)
Y = sigmoid(X)
plt.plot(X, Y)
plt.ylim(-0.1, 1.1)
plt.show()
```

```python
# compare

plt.plot(X, Y)
Y = step_function(X)
plt.plot(X, Y)
plt.show()
```

```python
# ReLU function
import matplotlib.pylab as plt
import numpy as np


def relu(x):
    return np.maximum(0, x)


x = np.arange(-5.0, 5.0, 0.1)
y = relu(x)
plt.plot(x, y)
plt.ylim(-1.0, 5.5)
plt.show()
```
