```python
import numpy as np

A = np.array([1, 2, 3])

print(A)
print(np.ndim(A))
print(A.shape)
print(A.shape[0])
```

```python
B = np.array([[1, 2], [3, 4], [5, 6]])
print(B)
print(np.ndim(B))
print(B.shape)
```

```python
print(np.dot(A, B))
# print(np.dot(B, A))  # ValueError: shapes (3,2) and (3,) not aligned: 2 (dim 1) != 3 (dim 0)
print(A @ B)
```
